```python
def flag(df):
    if (df['TIERED_NSF_OPTION_SET_NBR'] >= 0) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 4):
        return 0
    if (df['TIERED_NSF_OPTION_SET_NBR'] > 4) and (df['TIERED_NSF_OPTION_SET_NBR'] == 5):
        return 1
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 24) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 250):
        return 2
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 274) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 500):
        return 3
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 524) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 750):
        return 4
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 774) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 1000):
        return 5
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 1024                                                                                                                                                                                                                                   ) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 1250):
        return 6
    elif (df['TIERED_NSF_OPTION_SET_NBR'] > 1274) and (df['TIERED_NSF_OPTION_SET_NBR'] <= 1500):
        return 7

df['TIER_LEVEL'] = df.apply(flag, axis = 1)
```
